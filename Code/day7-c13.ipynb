```python
import numpy as np
import matplotlib.pyplot as plt
#numpy->matplotlib
#dataframe->seaborn

%matplotlib inline 
#it only exits in jupyter notebook
#otherwise, use plt.show()

#plt.style.available[:2]
print(plt.style.available[:5])
```

```python
#plt.style.use(), plt.style.context()
plt.style.use("seaborn-whitegrid")
#plot curve
x = np.linspace(-5, 5, 100)
y1 = np.divide( 1, 1+np.exp(-x) ) #1/(1+exp(-x))
y2 = np.divide( np.exp(x)-np.exp(-x), np.exp(x)+np.exp(-x) ) #exp(x)-exp(-x) / exp(x)+exp(-x)
#set parameters for line and marker
plt.plot(x, y1, c="g", ls="-", lw=1, marker="o", ms= 2) #c for color, ls for linestyle, lw for linewidth
plt.plot(x, y2, c="b", ls="-", lw=1, marker="+", ms= 2)
#xlim(),ylim(), axis([])
plt.xlim(-1,1)
#xticks([], fontsize)
plt.xticks(np.arange(-5, 6))
plt.tick_params(axis="both", labelsize=10) #labelsize for fontsize in xticks() and ysticks()
#xlabel,ylabel, title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("sigmoid & tanh", fontsize=15)
#plt.legend()设置图例
plt.legend(["sigmoid(x)", "tanh(x)"], loc="upper left", frameon=True, fontsize=15)
#plt.annotate()设置箭头及文字
plt.annotate("tanh(0)=0", xy=(0,0), xytext=(2,-0.5), fontsize=10, \
             arrowprops=dict(facecolor="yellow", edgecolor="yellow", shrink=0.1))
#plt.savefig(loc)
plt.savefig("sig&tanh.png")
```

```python
#scatter curve
x = np.linspace(-5, 5, 20)
y1 = np.divide( 1, 1+np.exp(-x) ) #1/(1+exp(-x))
#color according to y
plt.scatter(x,y1, c=y1, cmap="jet")
```

```python
#generate a random sequence
#the size of dot based on the seq
s_seq = np.random.randint(5, 100, size=20)
print(s_seq)
plt.scatter(x,y1, c=y1, s=s_seq, cmap="jet")
```

```python
#set alpha=0.2 透明度
plt.scatter(x,y1, c=y1, s=s_seq, cmap="jet", alpha=0.2)
```

```python
#set alpha=0.8 透明度
plt.scatter(x,y1, c=y1, s=s_seq, cmap="jet", alpha=0.8)
```

```python
#bar figure for sigmoid
plt.bar(x, y1)
```

```python
#barh figure for tanh
y2 = np.divide( np.exp(x)-np.exp(-x), np.exp(x)+np.exp(-x) ) #exp(x)-exp(-x) / exp(x)+exp(-x)
plt.barh(x, y2)
#bottom 竖轴
```

```python
#stacked bar
plt.bar(x, y1, label="sigmoid")
plt.bar(x, y2, bottom=y1, label="tanh")
```

```python
#subplots
#sigmoid, tanh, sin, cos
y3 = np.sin(x)
y4 = np.cos(x)
#plt.subplots_adjust(wspace,hspace)
#plt.subplot(324)
grid = plt.GridSpec(2, 2, wspace=0.2, hspace=0.4)
plt.subplot(grid[0,0])
plt.plot(x, y1)
plt.ylabel("sigmoid(x)")

plt.subplot(grid[0,1])
plt.plot(x, y2, label="tanh(x)")
plt.ylabel("tanh(x)")

plt.subplot(grid[1,0])
plt.plot(x, y3, label="sin(x)")
plt.ylabel("sin(x)")

plt.subplot(grid[1,1])
plt.plot(x, y4, label="cos(x)")
plt.ylabel("cos(x)")
```

```python
x_norm = np.random.randn(1000)
#frequency hist
plt.hist(x_norm, bins=50)
```

```python
#prob density hist
plt.hist(x_norm, bins=50, density=True)
```

```python
#cumulative prob hist
plt.hist(x_norm, bins=50, density=True, cumulative=True)
```

```python
#errorbar(yerr)
dy = 0.1
plt.errorbar(x, y1+dy*np.random.randn(20), yerr=0.1, fmt="+g")
```

```python
#object-oriented figure
#use ax
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax2 = fig.add_axes([0.3, 0.5, 0.2, 0.2])
#add_axes at one time
ax1.plot(x,y1)
ax2.plot(x,y2)
```

```python
#3d curve
from mpl_toolkits import mplot3d
t = np.linspace(0, 2*np.pi, 100)
x_3d = np.sin(t)
y_3d = np.cos(t)
z_3d = np.arange(t.size)
#plot3D, scatter3D, plot_surface()
ax = plt.axes(projection="3d")
ax.plot3D(x_3d, y_3d, z_3d)
```

```python
#plot_surface->z should be 2-dimensional
z_3d = np.arange(t.size).reshape(t.size,1)
ax = plt.axes(projection="3d") #get new ax, otherwise it doesn't work
ax.plot_surface(x_3d, y_3d, z_3d)
```
